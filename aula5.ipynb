```python
# Prompt: pesquisar os top 10 paises com imigrantes para o canada.

top_10 = df.sort_values(by='Total',ascending=False).head(10)
```

# *Temas do Seaborn*

```python
import matplotlib.pyplot as plt
```

```python
sns.set_theme()
```

```python
ax = sns.barplot(data=top_10,y=top_10.index,x='Total',orient='h')
ax.set(
    title='Os dez países com mais imigrantes para o Canada',
    xlabel='Total de Imigrantes',
    ylabel=''
)
plt.show()
```

```python
def gerar_grafico_paleta(palete):
  fig,ax = plt.subplots(figsize=(8,4))
  ax = sns.barplot(data=top_10,y=top_10.index,x='Total',orient='h',palette=palete)
  ax.set_title('Os dez países com mais imigrantes para o Canadá',loc='left',fontsize=16)
  ax.set_xlabel('Total de Imigrantes',fontsize=14)
  ax.set_ylabel('')
  plt.show()
```

```python
sns.set_theme(style='darkgrid')
gerar_grafico_paleta('coolwarm')
```

```python
sns.set_theme(style='ticks')
gerar_grafico_paleta('Spectral')
```

```python
sns.set_theme(style='dark')
gerar_grafico_paleta('vlag')
```
